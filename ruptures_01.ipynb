```python

# change point detection
model = "l1"  # "l2", "rbf"
algo = rpt.Dynp(model=model, min_size=3, jump=5).fit(uk['sessions'].values.reshape(-1,1))
my_bkps = algo.predict(n_bkps=3)

# show results
rpt.show.display(uk['sessions'].values.reshape(-1,1), my_bkps)
plt.title('sessions')
plt.show()

print(my_bkps)

#-------

# change point detection
model = "l1"  # "l2", "rbf"
algo = rpt.Dynp(model=model, min_size=3, jump=5).fit(uk['searches'].values.reshape(-1,1))
my_bkps = algo.predict(n_bkps=3)

# show results
rpt.show.display(uk['searches'].values.reshape(-1,1), my_bkps)
plt.title('searches')
plt.show()

print(my_bkps)

#-------


# change point detection
model = "l1"  # "l2", "rbf"
algo = rpt.Dynp(model=model, min_size=3, jump=5).fit(uk['bookings'].values.reshape(-1,1))
my_bkps = algo.predict(n_bkps=3)

# show results
rpt.show.display(uk['bookings'].values.reshape(-1,1), my_bkps)
plt.title('bookings')
plt.show()

print(my_bkps)

#-------
```
